```python
import re
import unicodedata
import os
import glob

path = "aihubBboxDetect"
files = glob.glob(path + '/*')

label = ''
file_name = ''

file = open("gt.txt", "w", encoding='UTF-8-sig')

for f in files:
  num = re.findall(r"\d+", f)
  label = f.replace(str(num[0]), '')
  label = label.replace('aihubBboxDetect/','')
  label = label.replace('_', '')
  label = label[19:]
  label = label.replace('.jpg', '')
  label = unicodedata.normalize('NFC', label)

  fpath = f.replace('aihubBboxDetect/', '')
  fpath = unicodedata.normalize('NFC', fpath)
  print(fpath + '\t' + label + '\n')
  file.write(fpath + '\t' + label + '\n')
file.close()

```
